Diabetes **Prediction**

Import Library

```python
import pandas as pd
```

Import Dataframe

Data Source:  https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv

```python
diabetes=pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv")
```

```python
diabetes.head()
```

```python
diabetes.info()
```

```python
diabetes.isnull().sum()
```

```python
#summary stastics
diabetes.describe()
```

```python
diabetes.columns
```

Define Features and Target

```python
y=diabetes['diabetes']
x=diabetes[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
x_train
```

```python
y_train
```

Import Model

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=300)
```

```python
model.fit(x_train,y_train)
```

Predict Using Model

```python
y_pred=model.predict(x_test)
y_pred
```

Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
